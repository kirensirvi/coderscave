```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = ("/content/Hackathon_Ideal_Data.csv")
df = pd.read_csv(data)
```

```python
heat_map = df[['VALUE', 'QTY']].corr()
f, ax = plt.subplots(figsize = (8,4))
sns.heatmap(heat_map, vmax=0.5, square = True, annot = True)
```

```python
Line_plot = df.groupby('QTY')['VALUE'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(Line_plot['QTY'], Line_plot['VALUE'], marker='o', linestyle='--', color='pink')
plt.title('Line chart of quantity vs value')
plt.xlabel('QTY')
plt.ylabel('VALUE')
plt.grid(True)
plt.show()
```

```python
pd.crosstab(index = df['STORECODE'],
            columns = df['VALUE'],
            margins = True,
            normalize = 'columns')
```

```python
sns.set(style="darkgrid")
sns.regplot(x=df['QTY'], y=df['VALUE'])
```

```python
sns.distplot(df['VALUE'])
```

```python
sns.countplot(x='QTY',data=df)
```

```python
sns.boxplot(x='QTY',data=df)
```

```python
plt.figure(figsize=(10, 8))
sns.stripplot(x="GRP", y="VALUE", data=df)
plt.title('Strip Plot for Qty and Value')
plt.xlabel('Qty')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.stripplot(x="CMP", y="MBRD", data=df, color="pink")
plt.title('Strip Plot for CMP and MBRD')
plt.xlabel('CMP')
plt.ylabel('MBRD')
plt.show()
```
